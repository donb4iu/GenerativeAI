```python
import chromadb
chroma_client = chromadb.Client()
chroma_client.list_collections()
```

```python
from sentence_transformers import SentenceTransformer
#sentence1 = ["This is an example sentence", "Each sentence is converted"]
sentence1 = "This is a document about pineapple"
sentence2 = "This is a document about oranges"

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
#embeddings = model.encode(sentences)
embedding1 = model.encode(sentence1)
embedding2 = model.encode(sentence2)
print(embedding1)
print(embedding2)
```

```python
collection_name = "my_collection"
if collection_name in [c.name for c in chroma_client.list_collections()]:
    chroma_client.delete_collection(collection_name)
    print("--- deleted ---")
```

```python
collection = chroma_client.create_collection(name=collection_name)
```

```python
from typing import List
from sentence_transformers import SentenceTransformer

sentences = [sentence1, sentence2]
ids = ["id1", "id2"]
model_name = "all-MiniLM-L6-v2"
# Load the SentenceTransformer model
model = SentenceTransformer(model_name)
# Compute embeddings for the sentences and convert them to lists
embeddings = [model.encode(sentence).tolist() for sentence in sentences]
# Add the sentences to the collection with their embeddings
collection.add(
        documents=sentences,
        ids=ids,
        embeddings=embeddings
              )
```

```python
results = collection.query(
    query_texts=["This is a query document about hawaii"], # Chroma will embed this for you
    n_results=2 # how many results to return
)
print(results)
```
